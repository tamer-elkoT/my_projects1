```python
# We want to check is there an outliers in the Fare column 
# We will use the IQR method to find the outliers
# First, we need to calculate the IQR
q1 = data["Fare"].quantile(.25)
q3 = data["Fare"].quantile(.75)
# Then, we calculate the IQR
iqr = q3 - q1
# Now, we can calculate the lower and upper bounds for outliers
lower_whisker = q1 -1.5 * iqr
upper_whisker = q3 + 1.5 * iqr
# Now, we can check if there are any outliers in the Fare column
# If there are any values in the Fare column that are less than the lower whisker or greater
# than the upper whisker, then there are outliers in the Fare column
# if the fare is negative but the lower_whisker = 0
if lower_whisker < 0 :
    lower_whisker = 0
median = data["Fare"].median()
q1,q3,iqr,lower_whisker,upper_whisker,median

```

```python
import matplotlib.pyplot as plt
ax = data["Fare"].plot(kind="kde")
ax.set_xlim((-10,100))
ax.axvline(lower_whisker,linestyle = "--",color="red")
ax.axvline(upper_whisker,color="red")
ax.axvline(x=q1,color = "green")
ax.axvline(x = q3 ,color="green")
ax.axvline(x=median,color="yellow")
```

```python
filt_lower = data["Fare"] < lower_whisker
filt_upper = data["Fare"] > upper_whisker
filt = filt_lower | filt_upper
outer_indices = data[filt].index


```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
sns.boxenplot(x=data["Fare"])
plt.title("Fare Box Plot with outliers")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data.index, y=data['Fare'], color='r', label='Outliers')
plt.xlabel('Index')
plt.ylabel('Fare')
plt.title('Outliers in Fare')
plt.legend()
plt.show()
```
