# get the index of the outliers 

```python
for col in numerical_columns.columns:

    Q1 = numerical_columns[col].quantile(0.25)
    Q3 = numerical_columns[col].quantile(0.75)
    IQR = Q3 - Q1 

    lower_whisker = Q1 - 1.5*IQR
    upper_whisker = Q3 + 1.5*IQR

    outliers = numerical_columns[(numerical_columns[col] < lower_whisker) | (numerical_columns[col] > upper_whisker) ][col]
    print(f"Outliers in '{col}':")
    if not outliers.empty:
        print(outliers)
    else :
        print('No outliers found')
    
    print("-" * 40)
```

```python
def calculate_IQR(df,col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    IQR = q3 - q1
    return IQR 
```

```python
def lower_whisker(df, col, IQR_value):
    q1 = df[col].quantile(0.25)
    lower = q1 - 1.5 * IQR_value
    return lower
```

```python
def upper_whisker(df, col, IQR_value):
    q3 = df[col].quantile(0.75)
    upper = q3 + 1.5 * IQR_value
    return upper
```

```python
def detect_outliers(df, col):
    IQR_value = calculate_IQR(df, col)
    lower = lower_whisker(df, col, IQR_value)
    upper = upper_whisker(df, col, IQR_value)
    
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    return outliers
```

## drop the outliers that belong to the major class Alive

```python
def outliers_in_major_class(df, target_col="Patient_Status"):
    major_class = df[target_col].mode()[0]  # Get the major class, assumed to be 'Alive'
    
    # Initialize an empty list to store the outlier indices
    outlier_indices = []
    
    # Iterate over the numerical columns
    for col in df.select_dtypes(include='number').columns:  # Only numerical columns
        # Detect outliers for the current column
        outliers_df = detect_outliers(df, col)
        
        # Get the indices of the outliers
        outlier_indices.extend(outliers_df.index.tolist())
    
    # Remove duplicates from the outlier indices
    outlier_indices = list(set(outlier_indices))
    
    # Get the outliers rows from the original DataFrame
    outliers_df = df.iloc[outlier_indices]
    
    # Filter out rows where the Patient_Status is 'Dead' (drop 'Dead' class from the outliers)
    outliers_in_major_class_df = outliers_df[outliers_df[target_col] == major_class]
    
    return outliers_in_major_class_df

# Call the function to get the outliers from the major class
outliers_in_major_class_df = outliers_in_major_class(df, target_col='Patient_Status')

# Display the result
outliers_in_major_class_df
```

```python
# Drop the outliers that belong to the major class
df_cleaned = df.drop(outliers_in_major_class_df.index)
df_cleaned.head()
```

##  Upsampling MALE Class to a Fraction of FEMALE Class

```python
X["Gender"].value_counts()
```

```python
df_cleaned.head()
```

```python
from sklearn.utils import resample
from sklearn.model_selection import train_test_split


# Separate the classes based on 'Gender'
df_majority = df_cleaned[df_cleaned["Gender"] == "FEMALE"]
df_minority = df_cleaned[df_cleaned["Gender"] == "MALE"]

# Calculate the number of samples for the minority class (MALE)
n_samples_male = int(len(df_majority) * 0.90)  # 90% of the majority class size

# Upsample the minority class (MALE)
df_minority_upsampled = resample(
    df_minority,
    replace=True,        # Sample with replacement
    n_samples=n_samples_male,  # Target size for MALE
    random_state=42
)

# Combine the upsampled minority class with the majority class
df_balanced = pd.concat([df_majority, df_minority_upsampled])



# Separate features (X) and target (y)
X_balanced = df_balanced.drop('Patient_Status', axis=1)
y_balanced = df_balanced['Patient_Status']


# Perform train-test split with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X_balanced, y_balanced, test_size=0.2, random_state=42, stratify=y_balanced
)

```
